```python
# Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of word
text=("The cat sat on the mat. The cat ate her dinner. The mouse chased the cat and the dog chased the mouse")

# Create the wordcloud object
wordcloud = WordCloud(width=1790, height=1680, margin=10).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
from pytube import YouTube

# Specify the URL of the YouTube video
video_url = 'https://www.youtube.com/watch?v=x6rNtBBspdo'

# Create a YouTube object with the URL
yt = YouTube(video_url)

# Get the highest resolution Progressive Stream available
stream = yt.streams.get_highest_resolution()

# Download the video
stream.download()
```
